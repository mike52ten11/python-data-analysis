```python
import pandas as pd
csv_path = '用戶1~3.csv'
data = pd.read_csv(csv_path,encoding='ANSI')
data
```

```python
# 根據fcustomerid列對數據進行分組
grouped = data.groupby('fcustomerid')
grouped
```

```python
for name, group in grouped:
    print('name-->',name)
    print('group-->\n',group)
    print('=======================================================================================')
```

```python
import os
#建立資料夾
os.makedirs('Group_by_fcustomerid',exist_ok=True)
# 對每個分組應用函數,將數據寫入新的CSV文件
for name, group in grouped:
    
    group.to_csv(f'Group_by_fcustomerid/output_{name}.csv', index=False,encoding='ANSI')
    print(f'{name} 完成')
```
